```python
import backtester_delayed
import matplotlib.pyplot as plt
%matplotlib notebook

def runTest(startDate, endDate, currencies=None):
    balancer, date = backtester_delayed.doBacktest(startDate, endDate, currencies)
    
    plt.plot(balancer.worth, color='blue')
    plt.plot(balancer.buyandholdworth, color='red')
    plt.show()
    
```

```python
%matplotlib notebook
runTest('2017-01-01', '2017-12-31')
```
